```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        result ="positive"
    elif sentiment < 0:
        result="negetive"
    else:
        result = "neutral"
    return result
```

```python
#test FXN
get_sentiment("i wish i should do somewhat better next time")
```

```python
df['emotions']=df['text'].apply(get_sentiment)
```

###keyword Extraction
+Extract most commonest words per class of emotion

```python
from collections import Counter
```

```python
def extract_keywords(text,num=50):
    tokens =[ tok for tok in text.split()]
    most_common_tokens = Counter(tokens).most_common(num)
    return dict(most_common_tokens)
```

```python
#list of emotions
emotion_list = df['sentiment'].unique().tolist()
emotion_list
```

```python
happiness_list=df[df['sentiment'] == 'happiness']['clean_text'].tolist()
happiness_list
```

```python
#happiness Document
happiness_docx= ' '.join(happiness_list)
```

```python
happiness_docx
```

```python
#extract keywords
keywords_happiness=extract_keywords(happiness_docx)
```

```python
keywords_happiness
```

```python
#plot
def plot_most_common_words(mydict):
    df_01 = pd.DataFrame(mydict.items(),columns=['token','count'])
    plt.figure(figsize=(40,20))
    sns.barplot(x='token',y='count',data=df_01)
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_most_common_words(keywords_happiness)
```

```python
surprise_list = df[df['sentiment'] == 'surprise']['clean_text'].tolist()
#document
surprise_docx=' '.join(surprise_list)
#extract keyword
keyword_surprise=extract_keywords(surprise_docx)
```

```python
plot_most_common_words(keyword_surprise)
```

```python
### word cloud
from wordcloud import WordCloud
```

```python
def plot_wordcloud(docx):
    mywordcloud = WordCloud().generate(docx)
    plt.figure(figsize=(20,10))
    plt.imshow(mywordcloud,interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
#modelling
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB


#vectorizer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

#Metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
#split our dataset
from sklearn.model_selection import train_test_split
```

```python
#### build features from our text
```

```python
Xfeatures = df['clean_text']
ylabels = df['sentiment']
```

```python
Xfeatures
```

```python
ylabels
```

```python
cv = CountVectorizer()
X = cv.fit_transform(Xfeatures)
```

```python
#Get Features By Name
cv.get_feature_names()
```

```python
X
```

```python
#split Dataset
X_train,X_test,y_train,y_test =  train_test_split(X,ylabels,test_size=0.3,random_state=42)
```

Building Model

```python
nv_model = MultinomialNB()
nv_model.fit(X_train,y_train)
```

```python
# accuracy
nv_model.score(X_test,y_test)
```

```python
#predictions
y_pred_for_nv = nv_model.predict(X_test)
```

```python
y_pred_for_nv
```

```python
def predict_emotion(sample_text,model):
    myvect = cv.transform(sample_text).toarray()
    prediction = model.predict(myvect)
    pred_proba = model.predict_proba(myvect)
    pred_percentage_for_all = dict(zip(model.classes_,pred_proba[0]))
    print("prediction:{}, Prediction Score:{}".format(prediction[0],np.max(pred_proba)))
    return pred_percentage_for_all
```

```python
predict_emotion(sample_text,nv_model)
```

```python
predict_emotion(["she hates running all day"],nv_model)
```
